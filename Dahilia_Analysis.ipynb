```python
###let's merge on all shared columns between the 4 dataframes!

common_columns = set(financial_df.columns)
for df in [housing_df, income_df, poverty_df]:
    common_columns = common_columns.intersection(df.columns)
common_columns = list(common_columns)

merged_census_data = financial_df
for df in [housing_df, income_df, poverty_df]:
    merged_census_data = pd.merge(merged_census_data, df, on=common_columns, how='inner')

merged_census_data.sample(3)
```

```python
# make sure relevant columns are numeric. drop any nans
merged_census_data['percent_below_poverty_level'] = pd.to_numeric(merged_census_data['percent_below_poverty_level'], errors='coerce')
merged_census_data.dropna(subset=['percent_below_poverty_level'], inplace=True)

merged_census_data['families_median_income'] = pd.to_numeric(merged_census_data['families_median_income'], errors='coerce')
merged_census_data.dropna(subset=['families_median_income'], inplace=True)
```

### Analysis
Move those in need of help. These meet the following requirements, in order of importance:

- Boyoon: Zones at high risk of flooding (high CFLD values)
- Boyoon: Large clusters of families (high families_number values )
- Krystal: Lower income families (high families_received columns, high percent_below_poverty_level, low families_median_income)


Identify areas where those in need of help can be relocated. These meet the following requirements, in order of importance:
- Venetia: Zones at low risk of flooding (CFLD values)
- Venetia: Large areas of land available for construction (area), or many vacant housing units (vacant_housing_units)
- Krystal: Income levels of families there match the levels of those in need (families_received columns, percent_below_poverty_level, families_median_income)

(2c) Identify families with similar income levels

```python
# median of median income values
median_income_top_20 = top_20['families_median_income'].median()

# threshold for "high" vacant_housing_units values
high_vacant_threshold = merged_df['vacant_housing_units'].quantile(0.75)

# filters
filtered_df = merged_df[(merged_df['CFLD_RISKV'] < merged_df['CFLD_RISKV'].median()) &
                        (merged_df['vacant_housing_units'] > high_vacant_threshold) &
                        (merged_df['families_median_income'] >= median_income_top_20 - 5000) &
                        (merged_df['families_median_income'] <= median_income_top_20 + 5000)]

print(filtered_df[['geography', 'CFLD_RISKV', 'vacant_housing_units', 'families_median_income']])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = ['families_median_income', 'percent_below_poverty_level', 'CFLD_RISKV']
X = merged_df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k = 15
kmeans = KMeans(n_clusters=k, random_state=42)
merged_df['cluster'] = kmeans.fit_predict(X_scaled)

cluster_characteristics = merged_df.groupby('cluster')[features].mean()
print(cluster_characteristics)
```

 Identify zones at high risk of flooding (high CFLD values)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Separate the data into counties with nonzero risk values and counties with zero risk values
risky_counties = merged_df[merged_df['CFLD_RISKV'] > 0]
non_risky_counties = merged_df[merged_df['CFLD_RISKV'] == 0]

# Combine the data for non-risky counties into a single row
non_risky_counties_aggregated = pd.DataFrame({'COUNTY': ['Others'], 'CFLD_RISKV': [non_risky_counties['CFLD_RISKV'].sum()]})

# Concatenate risky counties and aggregated non-risky counties
combined_data = pd.concat([risky_counties, non_risky_counties_aggregated])


plt.figure(figsize=(12, 6))
sns.barplot(x='COUNTY', y = 'CFLD_RISKV', data = combined_data, palette='Reds')
plt.title('High Risk Coastal Flooding Values by County')
plt.xlabel('County')
plt.ylabel('Flooding Risk Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data for risk-existing counties
high_risk_counties = merged_df[merged_df['CFLD_RISKV'] > 0]

# Group by 'COUNTY' and calculate the total number of families for each filtered county above
county_total_families = high_risk_counties.groupby('COUNTY')['families_number'].sum().reset_index()

# Sort the high-risk counties by total number of families in descending order
county_total_families_sorted = county_total_families.sort_values(by='families_number', ascending=False)


plt.figure(figsize=(20, 8))
sns.barplot(x='COUNTY', y='families_number', data=county_total_families_sorted, palette='Blues')
plt.title('Total Number of Families in High-Risk Counties')
plt.xlabel('County')
plt.ylabel('Total Number of Families')
plt.xticks(rotation=90, ha='center')
plt.tight_layout()
plt.show()
```

```python
high_risk_counties = merged_df[merged_df['CFLD_RISKV'] > 0]
county_median_income = high_risk_counties.groupby('COUNTY')['families_median_income'].median().reset_index()

# the highest-risk counties/tracts sorted by lowest to highest families median income
county_median_income_sorted = county_median_income.sort_values(by='families_median_income')

# 20 counties with lowest families median income
counties_in_need = county_median_income_sorted.head(20)
counties_in_need=pd.merge(counties_in_need, high_risk_counties, on='COUNTY', how='inner')
counties_in_need
```

```python
counties_in_need=counties_in_need.drop(['families_median_income_y'],axis=1)
counties_in_need.rename(columns={'families_median_income_x': 'families_median_income'}, inplace=True)
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(x='COUNTY', y='families_median_income', data=counties_in_need, palette='Blues')
plt.title('Median Family Income ofTop 20 High-Need Counties')
plt.xlabel('County')
plt.ylabel('Median Income of Families')
plt.xticks(rotation=90, ha='center')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(x='COUNTY', y='families_median_income', data=counties_in_need, palette='Blues')
plt.title('Median Family Income of Top 20 High-Need Counties')
plt.xlabel('County')
plt.ylabel('Median Income of Families')
plt.xticks(rotation=90, ha='center')
plt.tight_layout()
plt.show()
```

```python
# median of median income values
median_income_top_20 = counties_in_need['families_median_income'].median()

# threshold for "high" vacant_housing_units values
high_vacant_threshold = merged_df['vacant_housing_units'].quantile(0.75)

# these are counties which have a lower risk value than the median risk value of counties in need, and also match the income level of counties in need
relocation_counties = merged_df[(merged_df['CFLD_RISKV'] < merged_df['CFLD_RISKV'].median()) &
                        (merged_df['vacant_housing_units'] > high_vacant_threshold) &
                        (merged_df['families_median_income'] >= median_income_top_20 - 5000) &
                        (merged_df['families_median_income'] <= median_income_top_20 + 5000)]

relocation_counties_least_risky = relocation_counties.sort_values(by='CFLD_RISKV')

relocation_counties_least_risky
```
